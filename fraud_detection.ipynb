so you can see this is a unbalanced dataset and this kind of makes logic,beacuse the number of fraud detection will be very less in the comparision to the non-frraud ones 

##### But there is a problem with this unbalanced dataset , if wo simply feed this dataset to the model then there is chances that the model will not learn about the fraud transactions patterns because they are very less in number comapred to non-fraud 

```python
# Separating the dataset for the transactions 
legit = credit_card_data[credit_card_data['Class'] == 0]
fraud = credit_card_data[credit_card_data['Class'] == 1]
```

```python
# Getting the statistical Measures for the fraud data
fraud['Amount'].describe()
```

```python
# Comparing the transsactions values
credit_card_data.groupby('Class').mean()
```

```python
# Building a sample dataset with same no of fraud and non-fruad sample thus balancing the dataset 
legit_sample = legit.sample(n=492)
```

```python
# concatinating the fraud and the legit_sample 
new_df = pd.concat([legit_sample , fraud], axis = 0)
# axis 0 -> rows and axis 1 -> cols
```

```python
new_df
```

```python
# Checking the distriobution of valid and non-valid tranasactions 
new_df['Class'].value_counts()
```

So now the dataset is Balanced 

```python
new_df.groupby('Class').describe()
```

#### Spliting for independent and Dependent data

```python
X = new_df.drop(columns = 'Class',inplace=False)
y = new_df['Class']
```

```python
X.shape
```

```python
y.shape
```

```python
# Import necessary libraries for PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Before performing the PCA we have t0o standardise the data

# Standardize the dataset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA to reduce the number of features
pca = PCA(n_components= 15)  # You can adjust n_components based on explained variance
X_pca = pca.fit_transform(X_scaled)
```

```python
# Check the explained variance ratio to see how much variance is retained
explained_variance = pca.explained_variance_ratio_
print(f'Explained variance by each component: {explained_variance}')
print(f'Total explained variance: {np.sum(explained_variance)*100:.2f}%')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, stratify= y,  random_state=2)
# stratify=y: This ensures that the class distribution in the target variable y remains the same in both the training and testing sets. This is particularly useful when working with imbalanced datasets.
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, y_train)
```

```python
# accuracy on training data
from sklearn.metrics import accuracy_score
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy*100)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy*100)
```

Even for a balanced dataset, it's important to look at precision and recall, especially in cases like fraud detection where the consequences of misclassification (false positives and false negatives) are critical.

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Predictions on test data
X_test_prediction_pca = model.predict(X_test)

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, X_test_prediction_pca)
recall = recall_score(y_test, X_test_prediction_pca)
f1 = f1_score(y_test, X_test_prediction_pca)

print(f'Precision: {precision*100:.2f}%')
print(f'Recall: {recall*100:.2f}%')
print(f'F1-Score: {f1*100:.2f}%')
```
